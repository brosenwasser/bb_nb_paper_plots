```python
import pandas as pd
import numpy as np
```

```python
#read in master table
master = pd.read_table('masterlist.txt', header=None,names=['index','wiyn_id','ra_hr','ra_min','ra.sec','deg','deg_min',
                                                           'deg_sec','ra_dec','deg_dec','wiyn_fiber','z'], 
                       delim_whitespace=True)
```

```python
#photometry table is three entries per single target, define these lists for each similar line
list1=list(range(0,736314,3))
list2=list(range(1,736314,3))
list3=list(range(2,736314,3))
```

```python
list12=sorted(list1+list2)
list23=sorted(list2+list3)
list13=sorted(list1+list3)
```

```python
#read in bb photometry table, skip lines with indices from other two lists
bb_data_1 = pd.read_table('ben_table_send',header=None,skiprows=list23,delim_whitespace=True,low_memory=False)
bb_data_2 = pd.read_table('ben_table_send',header=None,skiprows=list13,delim_whitespace=True,low_memory=False)
bb_data_3 = pd.read_table('ben_table_send',header=None,skiprows=list12,delim_whitespace=True,low_memory=False)
```

```python
bb_data_1.columns=['ra','dec','K','H','J','Y','Z','I','G']
bb_data_2.columns=['ra','dec','K','H','J','Y','Z','I','G']
bb_data_3.columns=['zcos','zdei','zwiyn']
```

```python
#delete columns not in use in each array
del bb_data_1['K']
del bb_data_1['H']
del bb_data_1['J']
del bb_data_2['ra']
del bb_data_2['dec']
del bb_data_2['Y']
del bb_data_2['Z']
del bb_data_2['I']
del bb_data_2['G']
```

```python
#combine all the arrays
bb_data_1 = pd.concat([bb_data_1,bb_data_2,bb_data_3],axis=1)
```

```python
#array for new columns for master
newcols = pd.DataFrame(0,index=list(range(0,665,1)),columns=['K','H','J','Y','Z','I','G','zcos','zdei','zwiyn'])
```

```python
master = pd.concat([master,newcols],axis=1)
```

```python
#match entries within 1.5 arcseconds
for i in range(645,665):
    for j in range(0,245437):
        if ((master.ra_dec[i]-bb_data_1.ra[j])**2 + (master.deg_dec[i]-bb_data_1.dec[j])**2)**0.5 <0.0004166:
            master.loc[i,'K'] = bb_data_1.loc[j,'K']
            master.loc[i,'H'] = bb_data_1.loc[j,'H']
            master.loc[i,'J'] = bb_data_1.loc[j,'J']
            master.loc[i,'Y'] = bb_data_1.loc[j,'Y']
            master.loc[i,'Z'] = bb_data_1.loc[j,'Z']
            master.loc[i,'I'] = bb_data_1.loc[j,'I']
            master.loc[i,'G'] = bb_data_1.loc[j,'G']
            master.loc[i,'zcos'] = bb_data_1.loc[j,'zcos']
            master.loc[i,'zdei'] = bb_data_1.loc[j,'zdei']
            master.loc[i,'zwiyn'] = bb_data_1.loc[j,'zwiyn']
```

```python
#read in nb table
nb_data = pd.read_table('ben_nb_table',header=0,delim_whitespace=True,low_memory=False)
nb_data
```

```python
nbcol = pd.DataFrame(0,index=list(range(0,665,1)),columns=['nb'])
master = pd.concat([master,nbcol],axis=1)
master
```

```python
for i in range(284,285):
    for j in range(0,127353):
        if ((master.ra_dec[i]-nb_data.ra[j])**2 + (master.deg_dec[i]-nb_data.dec[j])**2)**0.5 <0.0004166:
            master.loc[i,'Y'] = nb_data.loc[j,'y']
            master.loc[i,'Z'] = nb_data.loc[j,'z']
            master.loc[i,'I'] = nb_data.loc[j,'i']
            master.loc[i,'G'] = nb_data.loc[j,'g']
            master.loc[i,'zcos'] = nb_data.loc[j,'zcosmos']
            master.loc[i,'zdei'] = nb_data.loc[j,'deimos']
            master.loc[i,'zwiyn'] = nb_data.loc[j,'wiyn']
            master.loc[i,'nb'] = nb_data.loc[j,'nb']
            break
```
